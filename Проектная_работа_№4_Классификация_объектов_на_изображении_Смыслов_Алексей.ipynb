```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def find_coins(path: str):
    # Загрузка изображения
    img = cv2.imread(path)
    if img is None:
        print("Ошибка: изображение не загружено. Изображение не найдено или указан неверный путь")
        return

    # Преобразование в градации серого
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Улучшение контрастности с использованием CLAHE
    #clahe = cv2.createCLAHE(clipLimit=10.0, tileGridSize=(8, 8))
    #gray = clahe.apply(gray)

    # Размытие для уменьшения шума
    gray_blurred = cv2.medianBlur(gray, 5)

    # Параметры для функции HoughCircles
    dp = 1.2               # Параметр разрешения аккумуляторного массива
    minDist = 50           # Минимальное расстояние между центрами обнаруженных окружностей
    param1 = 50            # Порог для метода Кэнни
    param2 = 80           # Порог для функции HoughCircles
    minRadius = 50         # Минимальный радиус круга
    maxRadius = 100        # Максимальный радиус круга

    # Обнаружение кругов
    circles = cv2.HoughCircles(gray_blurred, cv2.HOUGH_GRADIENT, dp, minDist,
                               param1=param1, param2=param2,
                               minRadius=minRadius, maxRadius=maxRadius)

    img_circles = img.copy()
    total_amount = 0
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for circle in circles[0, :]:
            center = (circle[0], circle[1])
            radius = circle[2]
            # Рисование круга и центра
            cv2.circle(img_circles, center, radius, (0, 255, 0), 2)
            cv2.circle(img_circles, center, 5, (0, 0, 255), 3)

        for circle in circles[0, :]:
            radius = circle[2]
            if radius > 60 and radius < 68:
              nom = 1
            elif radius >= 68 and radius <= 72:
              nom = 10
            elif radius > 72 and radius < 80:
              nom = 2
            elif radius > 80:
              nom = 5
            total_amount += nom
            cv2.putText(img_circles, f"{nom} руб.",
                        (circle[0]-20, circle[1]-20),
                        cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)
            cv2.putText(img_circles, f"R={radius}",
                        (circle[0]-20, circle[1]+40),
                        cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)
        print(f"Общая сумма монет: {total_amount}₽")
        cv2.putText(img_circles, f"Сумма: {total_amount} руб.",
                    (20, 70), cv2.FONT_HERSHEY_COMPLEX, 1.5, (255, 0, 0), 2)
    else:
        print("Круги не найдены")

    # Отображение результатов
    plt.figure(figsize=(18, 12))
    plt.subplot(1, 2, 1)
    plt.title('Градации серого с CLAHE')
    plt.imshow(gray_blurred, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title('Обнаруженные круги (монеты)')
    plt.imshow(cv2.cvtColor(img_circles, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np

def process_frame(frame, previous_circles, frame_count, update_interval=10):
    """
    Обрабатывает кадр для определения монет и их номиналов.

    Параметры:
    - frame: numpy.ndarray — текущий кадр видео.
    - previous_circles: numpy.ndarray | None — координаты кругов, определенные ранее.
    - frame_count: int — текущий номер кадра.
    - update_interval: int — интервал обновления кругов.

    Возвращает:
    - circles: numpy.ndarray | None — координаты найденных кругов (если есть).
    """
    # Используем предыдущее значение кругов, если не обновляем
    if frame_count % update_interval != 0 and previous_circles is not None:
        return previous_circles

    # Преобразование кадра в градации серого
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Увеличение контраста и размытие для уменьшения шума
    low_contrast_image = cv2.convertScaleAbs(gray, alpha=1.8)
    gray_blurred = cv2.medianBlur(low_contrast_image, 5)

    # Параметры для метода Хафа
    dp = 1.2  # Разрешение аккумулятора
    minDist = 50  # Минимальное расстояние между центрами кругов
    param1 = 50  # Порог для метода Кэнни
    param2 = 100  # Порог для метода Хафа
    minRadius = 50  # Минимальный радиус круга
    maxRadius = 300  # Максимальный радиус круга

    # Обнаружение кругов
    circles = cv2.HoughCircles(
        gray_blurred, cv2.HOUGH_GRADIENT, dp, minDist,
        param1=param1, param2=param2,
        minRadius=minRadius, maxRadius=maxRadius
    )

    if circles is not None:
        circles = np.uint16(np.around(circles))  # Округление координат

    return circles

def annotate_frame(frame, circles):
    """
    Добавляет аннотации (номиналы монет, радиус, сумму) на кадр.

    Параметры:
    - frame: numpy.ndarray — текущий кадр видео.
    - circles: numpy.ndarray | None — координаты кругов.

    Возвращает:
    - frame: numpy.ndarray — кадр с аннотациями.
    """
    if circles is None:
        return frame  # Если кругов нет, возвращаем кадр без изменений

    total = []  # Список для хранения номиналов монет
    for circle in circles[0, :]:
        nominal = 0
        center = (circle[0], circle[1])
        radius = circle[2]

        # Рисование кругов и центра
        cv2.circle(frame, center, radius, (0, 255, 0), 2)  # Контур круга
        cv2.circle(frame, center, 2, (0, 0, 255), 3)  # Центр круга

        # Определение номинала монеты на основе радиуса
        if 60 < radius < 70:
            nominal = 1  # 1 рубль
        elif 70 <= radius < 80:
            nominal = 2  # 2 рубля
        elif radius >= 80:
            nominal = 5  # 5 рублей
        total.append(nominal)  # Добавляем номинал в список

        # Нанесение текста с номиналом и радиусом
        cv2.putText(frame, f"{nominal} руб.", (circle[0] - 20, circle[1] - 20),
                    cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)
        cv2.putText(frame, f"R={radius}", (circle[0] - 20, circle[1] + 40),
                    cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)

    # Отображение общей суммы
    cv2.putText(frame, f"Сумма: {sum(total)} руб.", (20, 70),
                cv2.FONT_HERSHEY_COMPLEX, 1.5, (255, 0, 0), 2)

    return frame

def main():
    """
    Главная функция для работы с веб-камерой и записи видео.

    Включает захват кадров с камеры, обработку кругов методом Хафа,
    добавление аннотаций и сохранение обработанного видео.

    Примечание:
    - Камера должна быть расположена на расстоянии ~15 см от монет
      для корректного определения номиналов.

    """
    cap = cv2.VideoCapture(0)  # Открытие веб-камеры

    if not cap.isOpened():
        print("Не удалось открыть веб-камеру.")
        return

    # Параметры записи видео
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS)) or 30
    out = cv2.VideoWriter(
        'coin_detection_video.avi',
        cv2.VideoWriter_fourcc(*'XVID'),
        fps,
        (frame_width, frame_height)
    )

    print("Нажмите 'q' для завершения записи.")

    previous_circles = None
    frame_count = 0
    update_interval = 10  # Частота обновления кругов (каждые 10 кадров)

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Не удалось получить кадр с веб-камеры.")
            break

        frame_count += 1
        circles = process_frame(frame, previous_circles, frame_count, update_interval)

        # Если круги обновились, сохраняем их
        if frame_count % update_interval == 0:
            previous_circles = circles

        # Аннотируем кадр
        processed_frame = annotate_frame(frame, previous_circles)
        out.write(processed_frame)

        cv2.imshow("Coin detection", processed_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    out.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
